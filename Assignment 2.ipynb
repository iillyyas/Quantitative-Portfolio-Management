# Section 1

````python
import pandas as pd
import numpy as np
import yfinance as yf

# Date range
start_date = "2015-01-01"
end_date = "2020-12-31"

# Tickers' name
tic = "META AMZN AAPL NFLX GOOG"

# Importing the stock data using the download function from the yf package
prices = (yf.download(tickers=tic,
start=start_date,
end=end_date)
.reset_index()
.rename(columns = {"Date": "date",
                   "Open": "open",
"High": "high",
"Low": "low",
"Close": "close",
"Adj Close": "adjusted",
"Volume": "volume"
})
)
prices.head()
````

# Section 2

We first keep the ```date``` and ```adjusted prices``` data from the ```Price``` dataframe. We store them in the dataframe ```return``` that we sort by dates. <br>
We then compute the returns from the adjusted prices and store it in the ```ret``` dataframe. We add a level above the names of the stock in the ret dataframe to have the same number of levels as with our original dataframe returns.
We finally merge the two together and keep everything in the ```returns``` dataframe.<br>

We then compute the first moment, second moment and covariance matrix and display the results.

````python
# Creating a new dataframe containing only the dates, and the adjusted prices
returns = prices[['date','adjusted']]
returns.sort_values("date") 

# Computing the returns from the adjusted prices and storing them in ret
ret = returns.adjusted.pct_change()

# Adding a level above the tickers called 'returns' to have the same number of levels as
# the returns dataframe
ret.columns = pd.MultiIndex.from_product([['returns'],ret.columns])

# Merging the two dataframes
returns = returns.join(ret)

returns.head()
````

````python
# Computing the moments
first_mom = returns['returns'].mean()
second_mom = returns['returns'].var()
covar = returns['returns'].cov()

# Displaying the results
print('The first moment of the returns is:')
print(first_mom.to_string(),'\n')

print('The variance of the returns is:')
print(second_mom.to_string(),'\n')

print('The covariance matrix of the returns is:')
print(covar.to_string())
````

# Section 3

We compute the skewness and the excess kurtosis, and display the results.<br>
Daily stock returns seem to **not have a normal distribution** because their skewness is different from zero, and their excess kurtosis is positive. This would have to be confirmed using significance tests.

````python
# Computing the moments
skewness = returns['returns'].skew()
excess_kurtosis = returns['returns'].kurtosis()-3


# Displaying the results
print('The skewness of the returns is:')
print(skewness.to_string(),'\n')

print('The excess kurtosis of the returns is:')
print(excess_kurtosis.to_string(),'\n')
````
